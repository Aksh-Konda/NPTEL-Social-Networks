## Lec46 Homophily : Fatman Evolutionary Model - Base code (Adding People)

```python
import networkx as nx
import matplotlib.pyplot as plt
import random as rd

def create_graph():
    G = nx.Graph()
    #G.add_nodes_from(range(1, 101))
    for i in range(1, 101):
        G.add_node(i)
    return G

def visualize(G, labeldict, nsize):
    #nx.draw(G, labels = labeldict, with_labels=1, node_size = nsize)
    nx.draw(G, with_labels=1, node_size = nsize)
    plt.show() 
G = create_graph()

#visualize(G)

```

```python
#Assigning BMI 
def assign_bmi(G):
    for each in G.nodes():
        G.node[each]['name'] = rd.randint(15, 40)
        G.node[each]['type'] = 'person'
        
        
def get_labels(G):
    dict1 = {}
    for each in G.nodes():
        dict1[each] = G.node[each]['name']
    return dict1

def get_size(G):
    l1 = []
    for each in G.nodes():
        l1.append(G.node[each]['name'] *5)#Scaling for beautiful G
    return l1

G = create_graph()
assign_bmi(G)

label_dict = get_labels(G)
print labeldict

node_size = get_size(G)

visualize(G, labeldict, node_size)
```
